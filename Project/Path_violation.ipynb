```python
# Import Splinter, BeautifulSoup, Regular Expression, and Time
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import time
import wget
```

```python
executable_path = {'executable_path': 'C:/Users/renie/Desktop/chromedriver.exe'}

browser = Browser('chrome', **executable_path)
```

```python
months = ["january","march","april","may","june","july","august","september",\
          "october","november","december"]

download_directory = "2019/"
csv_files = []

for month in months:
    url = "https://opendata.dc.gov/datasets/moving-violations-issued-in-" + month + "-2019/data"
    #print(url)
    browser.visit(url)
    time.sleep(2)

    browser.is_element_present_by_text("Spreadsheet", wait_time=1)
    html = browser.html
    violations_soup = bs(html, 'html.parser')
    #violations_soup
    
    link_info = violations_soup.find("a", id="ember104")
    link = link_info['href']
    filename = link_info['download']
    
    csv_files.append(filename)
    print(link)
    print(filename)
#     print()

    download = wget.download(link, out=download_directory)
    print(download)
    
    time.sleep(10)
```

```python
column_names = ["ISSUE_DATE","VIOLATION_PROCESS_DESC","DRV_LIC_STATE","FINE_AMOUNT", "ISSUING_AGENCY_NAME", "LOCATION"]

all_data_twozeroonenine_df = pd.DataFrame(columns = column_names)
print(all_data_twozeroonenine_df)

first_month = "True"

months = ["January","March","April","May","June","July","August","September",\
          "October","November","December"]

csv_files = []

for month in months:
    filename = "Moving_Violations_Issued_in_" + month + "_2019.csv"
    csv_files.append(filename)

# Make sure the sub-folder name is 2019; otherwise, change it.
for file in csv_files:
    file_path = "2019/" + file
    print(file_path)
    
    month_df = pd.read_csv(file_path)
    
    temp_df = month_df.loc[:,["ISSUE_DATE","VIOLATION_PROCESS_DESC","DRV_LIC_STATE","FINE_AMOUNT", "ISSUING_AGENCY_NAME", "LOCATION"]]
    #temp_df = temp_df[temp_df['DRV_LIC_STATE'].isin(["DC","MD","VA"])==True]
    #temp_df.dropna(subset=["DRV_LIC_STATE"]) 
    
    if first_month == "True":
        all_data_twozeroonenine_df = temp_df 
        first_month = "False"
    else:    
        csv_df = pd.concat([all_data_twozeroonenine_df, temp_df])
        all_data_twozeroonenine_df = csv_df
        #break
    
all_data_twozeroonenine_df.to_csv("2019/final.csv")      
```

```python

all_data_twozeroonenine_df["YEAR"] = all_data_twozeroonenine_df['ISSUE_DATE'].str.split('/').str[0]
all_data_twozeroonenine_df["MONTH"] = all_data_twozeroonenine_df['ISSUE_DATE'].str.split('/').str[1]

all_data_twozeroonenine_df.sample(5)
```

### Question 6 - Total amount of issuing violations by State and Month
